```python

%%manim -qh -v WARNING Rigo
class Rigo(Scene):
  def construct(self):
      #Crear Grid de pruebas
      '''
      grid = NumberPlane()
      grid.add_coordinates()
      self.play(Write(grid))
      '''
      #Crear lineas de refenrencia
      Linea1 = Line([-1, 1, 0],[1, -1, 0])
      Linea2 = Line([-1, 1, 0],[-1, -1, 0])
      Linea3 = Line([-1, -1, 0],[1, -1, 0])

      #Sub Crear las llaves referenciadas a las lineas
      Llave1 = Brace(Linea1, direction=Linea1.copy().rotate(PI/2).get_unit_vector())
      TextoLlave1 = Llave1.get_text("$2\sqrt2$")

      Llave2 = Brace(Linea2, direction=Linea2.copy().rotate(PI/2).get_unit_vector())
      Llave2.rotate(PI, axis = Y_AXIS)
      Llave2.shift(0.6*LEFT)
      TextoLlave2 = Llave2.get_text("2")
      

      Llave3 = Brace(Linea3, direction=Linea3.copy().rotate(PI/2).get_unit_vector())
      Llave3.rotate(PI, axis = X_AXIS)
      Llave3.shift(0.75*DOWN)
      TextoLlave3 = Llave3.get_text("2")
      
      #Crear Titulo
      Titulo = Tex("Teorema de Pitgoras")
      Titulo.scale(1.4)
      Titulo.to_edge(ORIGIN)
      self.play(Write(Titulo), run_time = 2)
      self.play(
            Titulo.animate.scale(0.7).move_to(3.3*UP),
            run_time=2
        )
      #Crear un cuadrado

      Lado1 = 2.8284271247461900976033774484194
      Cuadrado1 = Square(
          side_length = Lado1,
          color = RED,
          fill_opacity = 1/2
      )
      Cuadrado1.rotate(45*DEGREES)
      Cuadrado1.shift(RIGHT)
      Cuadrado1.shift(UP)
      #Crear cuadrado 2

      Cuadrado2 = Square(
          side_length = 2,
          color = RED,
          fill_opacity = 1/2
      )
      Cuadrado2.shift(2*LEFT)
      #Crear Cuadrado 3
      Cuadrado3 = Square(
          side_length = 2,
          color = RED,
          fill_opacity = 1/2
      )
      Cuadrado3.shift(2*DOWN)
      

      
      
      v1 = [-1, -1, 0]
      v2 = [-1, 1, 0]
      v3 = [1, -1, 0]
      vertices = [v1, v2, v3]
      triangle = Polygon(*vertices, color=BLUE, fill_opacity=0.5)
      #Crear un grid
      grid = NumberPlane()
      grid.add_coordinates()
      self.play(Write(grid))
      #Crear triangulo rectangulo
      self.play(Write(triangle))
      self.play(Titulo.animate.move_to([-4.3, 3.3, 0]))
      #Agregar llaves de longitud & Agregar cuadrados correspondientes
      self.play(Write(Llave1))
      self.play(Write(TextoLlave1))
      self.wait()
      self.play(TextoLlave1.animate.move_to([3.7, 2.7, 0]))
      self.play(Llave1.animate.move_to([2.2, 2.2, 0]))
      self.play(Write(Cuadrado1))

      self.play(Write(Llave2))
      self.play(Write(TextoLlave2))
      self.wait()
      self.play(TextoLlave2.animate.move_to([-4.1, 0, 0]))
      self.play(Llave2.animate.move_to([-3.5, 0, 0]))
      self.play(Write(Cuadrado2))

      self.play(Write(Llave3))
      self.play(Write(TextoLlave3))
      self.wait()
      self.play(TextoLlave3.animate.move_to([0, -3.7, 0]))
      self.play(Llave3.animate.move_to([0, -3.4, 0]))
      self.play(Write(Cuadrado3))

      self.play(Unwrite(grid), run_time = 3.5)
      #Crear formula
      Formula = MathTex("a^2+b^2=c^2")
      Formula.move_to([4.5, 0.5, 0])

      Formula2 = MathTex("a^2+b^2=c^2")
      Formula2.move_to([4.5, -0.5, 0])

      Teorema = MathTex("(2)^2+(2)^2=(2\sqrt2)^2").move_to([4.5, -0.5, 0])
      Teorema2 = MathTex("2^2+2^2=(2\sqrt2)^2").move_to([4.5, -0.5, 0])
      Teorema3 = MathTex("4+4=8").move_to([4.5, -0.5, 0])
      Teorema4 = MathTex("8=8").move_to([4.5, -0.5, 0])

      self.wait()
      self.play(Write(Formula))

      self.play(Create(Formula2))
      self.wait()
      self.play(Transform(Formula2, Teorema))
      self.wait()
      self.play(Transform(Formula2, Teorema2))
      self.wait()
      self.play(Transform(Formula2, Teorema3))
      self.wait()
      self.play(Transform(Formula2, Teorema4))
      self.wait(2)

      #Des-Escribir TODO
      
      GrupoGen = VGroup(
          Titulo, 
          triangle,
          Cuadrado1,
          Cuadrado2,
          Cuadrado3,
          TextoLlave1, 
          TextoLlave2, 
          TextoLlave3,
          Llave1,
          Llave2,
          Llave3,
          Formula,
          Formula2
      )
      self.play(Unwrite(GrupoGen))
      self.wait(3)
      self.play(Write(Tex("Siwe Chicharron")))
      self.wait()

      

      
```

```python

%%manim -ql -v WARNING Llaves

class Llaves(Scene):
    def construct(self):
        A = Tex("Uno").move_to([0, 2, 0])
        B = Tex("Dos").move_to([0, 2, 0])
        C = Tex("Tres").move_to([0, 2, 0])
        D = Tex("Cuatro").move_to([0, 2, 0])
        self.wait()
        self.play(Write(A))
        self.wait()
        self.play(Transform(A, B))
        self.wait()
        self.play(Transform(A, C))
        self.wait()
        self.play(Transform(A, D))
        self.wait()
```

```python
%%manim -qm -v WARNING forms
class forms(Scene):
  def construct(self):
    t = MathTex("z=\\frac{x^2+y}{a^3}").scale(5)
    self.play(Write(t), run_time = 3.5)
    self.wait(3.4)
    self.play(Unwrite(t), run_time = 1.5)

```
